Основные значения:

* CRIM: This is the per capita crime rate by town _`(Уровень преступности на душу населения по городу)`_


* ZN: This is the proportion of residential land zoned for lots larger than 25,000 sq.ft. _`(Доля жилой земли, зонированной на участки больше 25 кв.футов)`_


* INDUS: This is the proportion of non-retail business acres per town. _`(Соотношение нерозничного бизнеса в акрах в городе)`_


* CHAS: This is the Charles River dummy variable (this is equal to 1 if tract bounds river; 0 otherwise) _`(Фиктивная переменная р.Чарльз (она равняется 1, если территория ограничена рекой, 0 - в противном случае))`_


* NOX: This is the nitric oxides concentration (parts per 10 million) _`(Концентрация оксидов азота (частей на 10 миллионов))`_


* RM: This is the average number of rooms per dwelling _`(Среднее количество комнат в жилище)`_


* AGE: This is the proportion of owner-occupied units built prior to 1940 _`(Соотношение занятых владельцами объектов, построенных до 1940 г.)`_


* DIS: This is the weighted distances to five Boston employment centers _`(Взвешенные расстояния до 5-ти Бостонских центроб занятости)`_


* RAD: This is the index of accessibility to radial highways _`(индекс доступности радиальных магистралей)`_


* TAX: This is the full-value property-tax rate per 10,000 dollars _`(Полная величина налога на имущество на 10 долларов)`_


* PTRATIO: This is the pupil-teacher ratio by town _`(Соотношение учеников и учителей по городам)`_


* B: This is calculated as 1000(Bk — 0.63)², where Bk is the proportion of people of African American descent by town _`(Рассчитывается как 1000*(Bk-0.63)², где Bk - соотношение людей афроамериканского происхождения в городе)`_


* LSTAT: This is the percentage lower status of the population _`(Процент населения с низким статусом)`_


* MEDV: This is the median value of owner-occupied homes in 1000 dollars _`(Средняя стоимость занятых собственниками домов, в 1000 долларов)`_

*__1.4. Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT.__*

* Разделить данные на 80% тренировочных, 20% тестовых
* Составить модель линейной регрессии
* Посчитать ошибку модели на тестовых данных
* Сделать точечный график данных и полученной прямой линейной регрессии

Преобразуем данные типа Series в Numpy array

```python
LSTAT_array = np.array(data.LSTAT).reshape(-1,1)
MEDV_array = np.array(data.MEDV).reshape(-1,1)
plt.plot(LSTAT_array, MEDV_array, 'r.')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Разделим данные на 80% тренировочных и 20% тестовых.

```python
from sklearn.model_selection import train_test_split

X_train, x_test, Y_train, y_test = train_test_split(
    LSTAT_array,
    MEDV_array,
    test_size = 0.2
)
```

Обучим модель на тренировочных данных.

```python
model.fit(X_train, Y_train)
```

Оценим погрешность модели.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

predicted_y = model.predict(x_test)
print(f'Среднеквадратичное отклонение: {mean_squared_error(predicted_y, y_test)}')
print(f'Средняя абсолютная погрешность: {mean_absolute_error(predicted_y, y_test)}')
```

4,57% - абсолютная погрешность модели на тестовых данных.

```python
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.title('Линейная регрессия от одной переменной')

x = np.linspace(0, 100, 200)
k = model.coef_
b = model.intercept_
y = k*x + b
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, Y_train, 'r.')
plt.plot(x_test, y_test, 'go')
plt.xlim(0,50)
plt.ylim(0,60)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

### Задача 3 (Дискретное распределение случайной величины)

*__3.1. Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?__*

```python
import random
    
def cube_experiment(num_exp, sum_elem = False):
    exp_results = [random.randint(1,6) for i in range(num_exp)]
    return exp_results

def hist_creation(exp_results,x_min, x_max):
    plt.hist(x = exp_results, bins=50)
    plt.xlim(x_min,x_max)
    plt.xlabel('Значение параметра')
    plt.ylabel('Частота')
    plt.title('Моделирование подбрасывания кубика')
```

```python
hist_creation(cube_experiment(10), 0, 7)
```

При малом количестве экспериментов (подбрасываний) нет определённого распределения.

```python
hist_creation(cube_experiment(10000), 0, 7)
```

При большом количестве экспериментов (подбрасываний) - равномерное распределение.

*__3.2. Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?__*

```python
hist_creation([sum(cube_experiment(100)) for i in range(1000)],0, 600)
```

$X_{1},...X_{n_{1}}$ - последовательность независимых случайных величин ($n_{1}=100$);


$S_{n_{1}}=\sum \limits _{i=1}^{n_{1}}X_{i}$ - сумма элементов последовательности.

Согласно ЦПТ сумма $n_{1}$ независимых одинаково распределённых случайных величин имеет распределение, близкое к нормальному при бесконечно большой выборке последовательностей $n_{2}\to \infty$ ($n_{2}=1000$).
