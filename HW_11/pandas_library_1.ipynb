## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
dataset = pd.read_csv('power.csv')
dataset.head()
```

```python
dataset[ dataset['country'].str.contains('lat|lit|est', case=False) ]['country'].unique()
```

```python
dataset_country = dataset[dataset.country.isin(['Latvia','Lithuania','Estonia'])]
dataset_country_category = dataset_country[dataset_country.category.isin([4,12,21])]
dataset_country_category_year = dataset_country_category[(dataset_country_category.year >= 2005) & (dataset_country_category.year <= 2010)]
result = sum(dataset_country_category_year[dataset_country_category_year.quantity > 0]['quantity'])
print(f'Суммарное потребление стран Прибалтики категорий 4, 12 и 21 за период с 2005 по 2010 года: {result}')
```

```python
result = sum(dataset[dataset.country.isin(['Latvia','Lithuania','Estonia']) & dataset.category.isin([4,12,21]) \
        & (dataset.year >= 2005) & (dataset.year <= 2010) & (dataset.quantity > 0)]['quantity'])
print(f'Суммарное потребление стран Прибалтики категорий 4, 12 и 21 за период с 2005 по 2010 года: {result}')
```
