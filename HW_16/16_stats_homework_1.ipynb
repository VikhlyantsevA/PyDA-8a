## Задание 1

Изобразите с помощью matplotlib. Изобразите x + 2y + 3z = 0.

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

def z_function(x, y):
    return -x / 3 - 2 * y / 3

x = np.linspace(0, 10, 20)
y = np.linspace(0, 10, 20)

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap=cm.coolwarm, edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(azim=50)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

Найдите угол между векторами x, y и x, z. 

```python
x = np.array([1,1])
y = np.array([2, 0])
z = np.array([0,2])
```

```python
def find_angle(vector_1, vector_2):
    """
    Подсчет косинуса угла между векторами a, b по их координатам
    """
    # длины векторов
    vector_1_Length = np.linalg.norm(vector_1)
    vector_2_Length = np.linalg.norm(vector_2)
    
    return np.dot(vector_1, vector_2) / (vector_1_Length * vector_2_Length)
```

```python
find_angle(x, y)
```

```python
find_angle(x, z)
```

Изобразите с помощью matplotlib. Изобразите вектор x + 2y + 3z.

```python
def show_vector(vector):
    ax = plt.axes()
    plt.xlim([0, vector[0]+1])
    plt.ylim([0, vector[1]+1])
    plt.xlabel('1st coordinate')
    plt.ylabel('2nd coordinate')
    plt.grid()
    ax.arrow( 0, 0, vector[0], vector[1], head_width=0.1, head_length=0.2, fc='k', ec='k' )
    plt.show()
```

```python
show_vector(x + 2*y+3*z)
```

## Задание 2

Найдите собственные значения и собственные вектора матриц (необходимо решение на numpy и решение по алгоритму на бумажке). Для матрицы 3x3 можно посмотреть на корни характеристического многочлена, посчитанные в numpy.

```python
m1 = np.array([[2, 2],
               [1, 3]])

w, v = np.linalg.eig(m1)

r1 = np.linalg.matrix_rank(m1)
print(r1)

for i in range(len(w)):
    print("Собственное значение " + str(w[i]))
    print("соответствующий ему собственный вектор " + str(v[:,i]))
```

```python
m2 = np.array([[4, 1, -1],
               [1, 4, -1],
               [-1, -1, 4]])

w, v = np.linalg.eig(m2)

r2 = np.linalg.matrix_rank(m2)
print(r2)

for i in range(len(w)):
    print("Собственное значение " + str(w[i]))
    print("соответствующий ему собственный вектор " + str(v[:,i]))
```
